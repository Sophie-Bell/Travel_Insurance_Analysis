#### Import Libraries

```python
import numpy as np
import pandas as pd
```

#### Load Data

```python
df = pd.read_csv('../data/raw/TravelInsurancePrediction.csv')

df.head(5)
```

#### Data Cleaning

```python
df.shape
```

```python
df.isnull().any()
```

```python
df.columns
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df.head(5)
```

```python
def convert_yes_no_to_binary(df, columns):
    for column in columns:
        df[column] = df[column].map({'Yes': 1, 'No': 0})
    return df

yes_no_columns = ['FrequentFlyer', 'EverTravelledAbroad','GraduateOrNot']
df = convert_yes_no_to_binary(df, yes_no_columns)
```

```python
df.head(5)
```

```python
df['Employment Type'] = df['Employment Type'].replace('Private Sector/Self Employed', 'Private Sector')

df.head(3)
```
